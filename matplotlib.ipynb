# Setup
Create randomized data to use in the examples

The documentation for Matlibplot is [here](https://www.tutorialspoint.com/matplotlib/index.htm).

```python
import random
player_count:int = 4
rounds_limit:int = 10
score_data :[dict] =[]
player_scores : [int] = [0 for i in range(player_count)]

for round_num in range(rounds_limit):
    new_round_scores = {"Round": round_num}
    for player_num in range(player_count):
        player_scores[player_num] += random.randint(1,6) + player_num
        new_round_scores[f"Player {player_num}"] = player_scores[player_num]
    score_data.append(new_round_scores)
print(score_data)

```

```python
from pandas import DataFrame

score_df = DataFrame(score_data)
score_df
```

# Line Graph
Line graph example with legend and labels.

```python
from matplotlib import pyplot as plt
legend_labels = []
for p in range(player_count):
    plt.plot(score_df["Round"], score_df[f"Player {p}"])
    legend_labels.append(f"Player {p}")
plt.legend(legend_labels)
plt.xlabel("Rounds")
plt.ylabel("Scores")
plt.show()
```

# Bar Graph
Use Matplotlib to create a bar graph. The documentation for this feature is here: https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = [f"Player {i}" for i in range(player_count)]
students = player_scores
ax.bar(langs,students)
plt.show()
```
